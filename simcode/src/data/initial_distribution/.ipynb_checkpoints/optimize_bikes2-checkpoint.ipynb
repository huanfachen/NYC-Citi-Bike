```python
import time

# Third-party libs.
import numpy as np
import matplotlib.pyplot as plt

# App libs.
import simcode.src.nycbike as nycbike
```

```python
NUM_STATIONS = 767  # based on Jan 2018 Citi Bike data
```

**Iteratively move one bike from station with longest bike idle time to station with smallest bike idle time.**

```python
startTime = time.time()

# Initial value for iteration.
currDistr = nycbike.BikeSharingSimulation().almostUniformWithTotalSum(
    NUM_STATIONS, nycbike.NUM_BIKES)

numIters = 5000
revenues = []
totalPickupWaitTimes = []
totalDropoffWaitTimes = []
totalIdleTimes = []
distributions = []
bestRevenue = -float('inf')
bestDistr = None

for iterNum in range(numIters):
    print '.',
    if (iterNum + 1) % int(numIters / 10) == 0:
        print (iterNum + 1),

    # Evaluate current distribution.
    stats = nycbike.BikeSharingSimulation().run(
        tripDataDir='../trip_statistics/',
        initialDistribution=currDistr)
    # Record stats for this iteration.
    revenue = stats['Revenue']
    revenues.append(revenue)
    totalPickupWaitTimes.append(stats['TimeWaitForCycle'].sum())
    totalDropoffWaitTimes.append(stats['TimeWaitForDropoff'].sum())
    totalIdleTimes.append(stats['IdleTime'].sum())
    distributions.append(np.copy(currDistr))
    if revenue > bestRevenue:
        bestRevenue = revenue
        bestDistr = currDistr

    # Find station with smallest idle time and at least one empty rack (needs more bikes?)
    idle_times = stats['IdleTime']
    eligibleStationIds = np.argwhere(currDistr < nycbike.RACKS)
    eligibleStationsMore = idle_times[eligibleStationIds]
    smallestIdleTime = eligibleStationIds[np.argmin(eligibleStationsMore)]
    assert min(eligibleStationsMore) == idle_times[smallestIdleTime]

    # Find station with longest idle time and at least one bike (needs less bikes?)
    eligibleStationIds = np.argwhere(currDistr >= 1)
    eligibleStationsLess = idle_times[eligibleStationIds]
    longestIdleTime = eligibleStationIds[np.argmax(eligibleStationsLess)]
    assert max(eligibleStationsLess) == idle_times[longestIdleTime]

    # Update the distribution.
    transferAmount = 1
    currDistr[longestIdleTime] -= transferAmount
    currDistr[smallestIdleTime] += transferAmount

print 'done. took %.2f seconds' % (time.time() - startTime)
```

```python
plt.plot(revenues, marker='.', markersize=10)
plt.xlabel('iteration number')
plt.ylabel('revenue')
plt.show()
```

```python
plt.plot(totalPickupWaitTimes, marker='.', markersize=10, color='green')
plt.xlabel('iteration number')
plt.ylabel('total pickup wait time')
plt.show()
```

```python
plt.plot(totalDropoffWaitTimes, marker='.', markersize=10, color='red')
plt.xlabel('iteration number')
plt.ylabel('total dropoff wait time')
plt.show()
```

```python
plt.plot(totalIdleTimes, marker='.', markersize=10, color='red')
plt.xlabel('iteration number')
plt.ylabel('total idle time')
plt.show()
```

```python
print 'best revenue', max(revenues)
```

**Iteratively move one bike from station with longest bike idle time to station with smallest bike idle time(Non-Stochastic).**

```python
startTime = time.time()

# Initial value for iteration.
currDistr = nycbike.BikeSharingSimulation().almostUniformWithTotalSum(
    NUM_STATIONS, nycbike.NUM_BIKES)

numIters = 5000
revenues = []
totalPickupWaitTimes = []
totalDropoffWaitTimes = []
totalIdleTimes = []
distributions = []
bestRevenue = -float('inf')
bestDistr = None

for iterNum in range(numIters):
    print '.',
    if (iterNum + 1) % int(numIters / 10) == 0:
        print (iterNum + 1),

    # Evaluate current distribution.
    stats = nycbike.BikeSharingSimulation().run(
        tripDataDir='../trip_statistics/',
        initialDistribution=currDistr, rngSeed=10)
    # Record stats for this iteration.
    revenue = stats['Revenue']
    revenues.append(revenue)
    totalPickupWaitTimes.append(stats['TimeWaitForCycle'].sum())
    totalDropoffWaitTimes.append(stats['TimeWaitForDropoff'].sum())
    totalIdleTimes.append(stats['IdleTime'].sum())
    distributions.append(np.copy(currDistr))
    if revenue > bestRevenue:
        bestRevenue = revenue
        bestDistr = currDistr

    # Find station with smallest idle time and at least one empty rack (needs more bikes?)
    idle_times = stats['IdleTime']
    eligibleStationIds = np.argwhere(currDistr < nycbike.RACKS)
    eligibleStationsMore = idle_times[eligibleStationIds]
    smallestIdleTime = eligibleStationIds[np.argmin(eligibleStationsMore)]
    assert min(eligibleStationsMore) == idle_times[smallestIdleTime]

    # Find station with longest idle time and at least one bike (needs less bikes?)
    eligibleStationIds = np.argwhere(currDistr >= 1)
    eligibleStationsLess = idle_times[eligibleStationIds]
    longestIdleTime = eligibleStationIds[np.argmax(eligibleStationsLess)]
    assert max(eligibleStationsLess) == idle_times[longestIdleTime]

    # Update the distribution.
    transferAmount = 1
    currDistr[longestIdleTime] -= transferAmount
    currDistr[smallestIdleTime] += transferAmount

print 'done. took %.2f seconds' % (time.time() - startTime)
```
