```python
import time

# Third-party libs.
import numpy as np
import matplotlib.pyplot as plt

# App libs.
import simcode.src.nycbike as nycbike
```

```python
NUM_STATIONS = 767  # based on Jan 2018 Citi Bike data
```

**Baseline: Uniform Distribution**

```python
numAveragingTrials = 50
revenues = np.zeros(numAveragingTrials)
uniformDistr = nycbike.BikeSharingSimulation().almostUniformWithTotalSum(
    NUM_STATIONS, nycbike.NUM_BIKES)

for i in range(numAveragingTrials):
    print '.',
    # Run the simulation once.
    stats = nycbike.BikeSharingSimulation().run(
        initialDistribution=uniformDistr,
        tripDataDir='../trip_statistics/')
    # Record revenue for this trial.
    revenues[i] = stats['Revenue']
```

```python
baseline_revenue = revenues.mean()
print revenues.mean()
print revenues.std()
```

**Iteratively assign bikes proportional to waiting times.**

```python
numIters = 100
currDistr = nycbike.BikeSharingSimulation().almostUniformWithTotalSum(
    NUM_STATIONS, nycbike.NUM_BIKES)
revenues = []
pickupWaitTimes = []
dropoffWaitTimes = []

for iterNum in range(numIters):
    # Run the simulation with current distribution.
    stats = nycbike.BikeSharingSimulation().run(
        initialDistribution=currDistr,
        tripDataDir='../trip_statistics/')
    print '.',
    revenues.append(stats['Revenue'])
    pickupWaitTimes.append(stats['TimeWaitForCycle'].sum())
    dropoffWaitTimes.append(stats['TimeWaitForDropoff'].sum())

    # Update distribution.
    ratios = stats['TimeWaitForCycle'] / stats['TimeWaitForCycle'].sum()
    currDistr = ratios * nycbike.NUM_BIKES
```

```python
plt.plot(revenues, marker='.', markersize=10)
plt.title('Iterative: Proportional to Wait Time')
plt.xlabel('iteration number')
plt.ylabel('revenue')
plt.show()
```

```python
plt.plot(pickupWaitTimes, marker='.', markersize=10, color='green')
plt.xlabel('iteration number')
plt.ylabel('total pickup wait time')
plt.show()
```

```python
plt.plot(dropoffWaitTimes, marker='.', markersize=10, color='red')
plt.xlabel('iteration number')
plt.ylabel('total dropoff wait time')
plt.show()
```

```python
print 'best revenue %f (iteration %d)' % (max(revenues), np.argmax(revenues))
```

**Iteratively move one bike from station with longest drop-off time to station with longest pick-up time.**

Note: sending station must have at least one bike, and receiving station must have at least one open rack.

```python
startTime = time.time()

# Initial value for iteration.
currDistr = nycbike.BikeSharingSimulation().almostUniformWithTotalSum(
    NUM_STATIONS, nycbike.NUM_BIKES)

numIters = 5000
revenues = []
totalPickupWaitTimes = []
totalDropoffWaitTimes = []
distributions = []
bestRevenue = -float('inf')
bestDistr = None

for iterNum in range(numIters):
    print '.',
    if (iterNum + 1) % int(numIters / 10) == 0:
        print (iterNum + 1),

    # Evaluate current distribution.
    stats = nycbike.BikeSharingSimulation().run(
        tripDataDir='../trip_statistics/',
        initialDistribution=currDistr)
    # Record stats for this iteration.
    revenue = stats['Revenue']
    revenues.append(revenue)
    totalPickupWaitTimes.append(stats['TimeWaitForCycle'].sum())
    totalDropoffWaitTimes.append(stats['TimeWaitForDropoff'].sum())
    distributions.append(np.copy(currDistr))
    if revenue > bestRevenue:
        bestRevenue = revenue
        bestDistr = currDistr

    # Find station with longest pickup wait time and at least one empty rack (needs more bikes?)
    pickup_wait_times = stats['TimeWaitForCycle']
    eligibleStationIds = np.argwhere(currDistr < nycbike.RACKS)
    eligiblePickupWaitTimes = pickup_wait_times[eligibleStationIds]
    longestPickupWaitId = eligibleStationIds[np.argmax(eligiblePickupWaitTimes)]
    assert max(eligiblePickupWaitTimes) == pickup_wait_times[longestPickupWaitId]

    # Find station with longest dropoff time and at least one bike (needs less bikes?)
    dropoff_wait_times = stats['TimeWaitForDropoff']
    eligibleStationIds = np.argwhere(currDistr >= 1)
    eligibleDropoffWaitTimes = dropoff_wait_times[eligibleStationIds]
    longestDropoffWaitId = eligibleStationIds[np.argmax(eligibleDropoffWaitTimes)]
    assert max(eligibleDropoffWaitTimes) == dropoff_wait_times[longestDropoffWaitId]

    # Update the distribution.
    transferAmount = 1
    currDistr[longestDropoffWaitId] -= transferAmount
    currDistr[longestPickupWaitId] += transferAmount

print 'done. took %.2f seconds' % (time.time() - startTime)
```

```python
plt.plot(revenues, marker='.', markersize=10)
plt.xlabel('iteration number')
plt.ylabel('revenue')
plt.show()
```

```python
plt.plot(totalPickupWaitTimes, marker='.', markersize=10, color='green')
plt.xlabel('iteration number')
plt.ylabel('total pickup wait time')
plt.show()
```

```python
plt.plot(totalDropoffWaitTimes, marker='.', markersize=10, color='red')
plt.xlabel('iteration number')
plt.ylabel('total dropoff wait time')
plt.show()
```
